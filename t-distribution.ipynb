```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python

# Define a range of x values
x = np.linspace(-4, 4, 1000)

# Standard normal distribution
normal_dist = stats.norm.pdf(x)

# T-distributions with different degrees of freedom
t_dist_df2 = stats.t.pdf(x, df=2)
t_dist_df5 = stats.t.pdf(x, df=5)
t_dist_df10 = stats.t.pdf(x, df=10)
t_dist_df30 = stats.t.pdf(x, df=30)

# Plotting
plt.plot(x, normal_dist, label='Normal Distribution', linestyle='--')
plt.plot(x, t_dist_df2, label='t-distribution df=2')
plt.plot(x, t_dist_df5, label='t-distribution df=5')
plt.plot(x, t_dist_df10, label='t-distribution df=10')
plt.plot(x, t_dist_df30, label='t-distribution df=30')

# Adding title and labels
plt.title('t-distributions vs. Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
df = 10
t_dist_df = stats.t.pdf(x, df=df)
plt.plot(x, normal_dist, label='Normal Distribution', linestyle='--')
plt.plot(x, t_dist_df, label=f't-distribution df={df}')
plt.legend()
plt.show()
```
