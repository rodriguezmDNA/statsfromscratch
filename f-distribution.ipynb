### f distribution

```python
from scipy.stats import f
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
palette = sns.color_palette()
```

```python
# Define the degrees of freedom for the numerator and denominator
df_between = 2  # degrees of freedom for the numerator
df_within = 147  # degrees of freedom for the denominator

# Define the value of F for which you want to calculate the CDF
f_value = 49.16004

# P value of observed F value
cdf_value = f.cdf(f_value, df_between, df_within)
1-cdf_value
```

```python
# Critical F value at alpha level
f.ppf(1-0.05,df_between,df_within)
```

```python
x_linspace = np.linspace(start=0,stop=5,num=1000)

for nd in [(1,1),(5,2),(20,20),(2,147)]:
    num,denom = nd
    pdf = f.pdf(x_linspace, num,denom)
    sns.lineplot(x=x_linspace,y=pdf, label = f'({num},{denom})')

plt.ylim(0,1.5)
plt.xlabel('X')
plt.ylabel('PDF')
plt.legend()
plt.show()
```

```python
df_between,df_within = 3,246
f_value = 3.5


cdf_value = f.cdf(f_value, df_between, df_within)
p_value = 1-cdf_value
p_value * 100
```

```python
f_critical = f.ppf(1-0.05,df_between,df_within)
f_critical
```

```python
x_linspace = np.linspace(start=0,stop=5,num=10000)
pdf = f.pdf(x_linspace, df_between,df_within)



plt.figure(figsize=(6,3))

ax = sns.lineplot(x=x_linspace,y=pdf, color='k' )

x, y = ax.lines[0].get_data()
plt.fill_between(x, y, where=(x >= f_value), color=palette[0], alpha=0.5)



plt.annotate(
    f'''F value={f_value}
The probability of observing this, 
or a more extreme F value is ~{p_value:.2%}
     ''', 
    size=8,
    ha='left',
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=.5'),
    xy=(f_value, 0.03), 
    xytext=(f_value, 0.25),
    arrowprops=dict(facecolor='white',width=3,headwidth=10)
)


plt.annotate(
    f'''At a significance level of 0.05\nThe critical F value is {f_critical:.2f}\n\nThis means that F values lower than this\nmight be observed 95% of the time.
     ''', 
    size=8,
    ha='left',
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=.5'),
    xy=(f_critical, 0.08), 
    xytext=(f_critical-1.5, 0.55),
    arrowprops=dict(facecolor='white',width=3,headwidth=10)
)



plt.title(f'F distribution ({df_between},{df_within})')

plt.ylim(0,0.85)
plt.xlabel('X')
plt.ylabel('PDF')
#plt.legend()
plt.show()
```
